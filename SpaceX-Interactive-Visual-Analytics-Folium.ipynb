# Task 2: Mark the success/failed launches for each site on the map


Next, let's try to enhance the map by adding the launch outcomes for each site, and see which sites have high success rates.
Recall that data frame spacex_df has detailed launch records, and the `class` column indicates if this launch was successful or not


```python
spacex_df.tail(10)
```

Next, let's create markers for all launch records. 
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

_TODO:_ Create a new column in `launch_sites` dataframe called `marker_color` to store the marker colors based on the `class` value


```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
```

```python
# Function to assign color to launch outcome
def assign_marker_color(launch_outcome):
    if launch_outcome == 1:
        return 'green'
    else:
        return 'red'
    
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)
spacex_df.tail(10)
```

_TODO:_ For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python
# Add marker_cluster to current site_map
# current site map
site_map = folium.Map(location=[28.563, -80.576], zoom_start=5)

# Marker Cluster
marker_cluster = MarkerCluster().add_to(site_map)

# Rocket markers (success/failure)
for index, record in spacex_df.iterrows():
    coord = [record['Lat'], record['Long']]
    marker_color = 'green' if record['class'] == 1 else 'red'
    outcome = "Success Landing" if record['class'] == 1 else "Failure Landing"
    
    folium.Marker(
        location=coord,
        icon=folium.Icon(color=marker_color, icon='rocket', prefix='fa'),
        popup=f"{record['Launch Site']}<br>{outcome}"
    ).add_to(marker_cluster)

# Add site name labels
for site, row in spacex_df.groupby('Launch Site').first().iterrows():
    folium.map.Marker(
        [row['Lat'], row['Long']],
        icon=folium.DivIcon(
            html=f"""<div style="font-size: 12pt; color: black;"><b>{site}</b></div>"""
        )
    ).add_to(site_map)

site_map
```

# TASK 3: Calculate the distances between a launch site to its proximities


Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.


You can calculate the distance between two points on the map based on their `Lat` and `Long` values using the following method:


```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# find coordinate of the closet coastline
# Example: Lat: 28.56367  Lon: -80.57163
# distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
launch_site_coord = [28.5623, -80.5774]  
# Example: Kennedy Space Center
coastline_coord = [28.56367, -80.57163]

distance_coastline = calculate_distance(
    launch_site_coord[0], launch_site_coord[1],
    coastline_coord[0], coastline_coord[1]
)
print(f"Distance to closest coastline: {distance_coastline:.2f} KM")
```

_TODO:_ After obtained its coordinate, create a `folium.Marker` to show the distance


```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# Launch site aur coastline coordinates
launch_site_coord = [28.5623, -80.5774]   # Kennedy Space Center
coastline_coord   = [28.56367, -80.57163]

# Distance calculate karo
distance_coastline = calculate_distance(
    launch_site_coord[0], launch_site_coord[1],
    coastline_coord[0], coastline_coord[1]
)

# Launch site marker
folium.Marker(
    launch_site_coord,
    popup="Launch Site",
    icon=folium.Icon(color='blue', icon='rocket')
).add_to(site_map)

# Coastline marker
folium.Marker(
    coastline_coord,
    popup="Closest Coastline",
    icon=folium.Icon(color='green', icon='cloud')
).add_to(site_map)

# Distance label marker
distance_marker = folium.Marker(
    coastline_coord,
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12px; color:#d35400;"><b>{:.2f} KM</b></div>'.format(distance_coastline),
    )
).add_to(site_map)

site_map
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
# Example launch site coordinate
launch_site_coord = [28.5623, -80.5774]   # Kennedy Space Center

# Example proximity coordinates (replace with actual MousePosition values)
city_coord     = [28.5383, -81.3792]   # Orlando
railway_coord  = [28.5721, -80.5853]   # Example nearby railway
highway_coord  = [28.5721, -80.6508]   # Example nearby highway
coastline_coord= [28.56367, -80.57163] # Closest coastline

# Dictionary of proximities
proximities = {
    "Closest City": city_coord,
    "Railway": railway_coord,
    "Highway": highway_coord,
    "Coastline": coastline_coord
}

# Loop through proximities
for name, coord in proximities.items():
    # Distance calculation
    distance = calculate_distance(
        launch_site_coord[0], launch_site_coord[1],
        coord[0], coord[1]
    )
    
    # Marker for proximity point
    folium.Marker(
        coord,
        popup=f"{name} ({distance:.2f} km)",
        icon=folium.Icon(color='green', icon='info-sign')
    ).add_to(site_map)
    
    # Distance label on map
    folium.Marker(
        coord,
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 12px; color:#d35400;"><b>{:.2f} KM</b></div>'.format(distance),
        )
    ).add_to(site_map)
    
    # Line between launch site and proximity
    folium.PolyLine(
        locations=[launch_site_coord, coord],
        weight=2,
        color='blue'
    ).add_to(site_map)

# Launch site marker
folium.Marker(
    launch_site_coord,
    popup="Launch Site",
    icon=folium.Icon(color='red', icon='rocket')
).add_to(site_map)

site_map
```

### 🚀 Launch Sites Proximity Analysis

After plotting distance lines between launch sites and nearby features (railways, highways, coastlines, and cities), we can observe the following:

---

#### 🚄 Are launch sites in close proximity to railways?  
✅ Yes. Launch sites are generally very close (within a few kilometers) to **railways**.  
👉 This is important because heavy rocket parts and fuel tanks are transported via rail, which is more efficient than roads.

---

#### 🛣️ Are launch sites in close proximity to highways?  
✅ Yes. Highways are also located near launch sites.  
👉 Highways provide easy access for **staff, equipment, and vehicles**, supporting daily logistics.

---

#### 🌊 Are launch sites in close proximity to coastlines?  
✅ Yes, almost all launch sites are located **right next to coastlines (often <1 km)**.  
👉 Launching rockets over the ocean reduces risk to populated areas and allows safe recovery of rocket stages in the sea.

---

#### 🏙️ Do launch sites keep certain distance away from cities?  
✅ Yes. Launch sites are usually **50–100 km or more away from cities**.  
👉 This ensures safety in case of accidents, preventing harm to large populations.

---

### 🌍 Findings Summary
- 🚀 Launch sites are **close to coastlines** → Safe trajectories over the ocean.  
- 🚄 They are **close to railways & highways** → Efficient transport & logistics.  
- 🏙️ They are **far from cities** → Safety and minimal risk to people.  


**Conclusion:**  
The strategic placement of launch sites balances **safety, logistics, and operational efficiency** by ensuring proximity to coastlines, highways, and railways, while keeping a safe distance from major cities.
